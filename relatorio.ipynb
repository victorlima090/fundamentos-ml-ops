# Sistema de recomendação de Vinhos


## Contexto
Neste projeto existe o seguinte cenário, uma grande empresa produtora de vinhos precisa que seus produtos sejam avaliados, e então, comercializados. Visando aumentar a popularidade da sua marca, a empresa gostaria de mandar para a avaliação de especialistas apenas os seus melhores vinhos, para que sua imagem fique atrelada a vinhos de alta qualidade.

Por isso, se faz necessário um sistema de classificação de vinhos, capaz de prever se um vinho terá uma nota alta ou baixa pelos avaliadores de renome. O sistema receberá uma lista com característica químicas da safra, e deverá classificar se ela é recomendada ou não para avaliação posterior.


## Objetivo
O objetivo desse projeto é criar uma aplicação que receberá os dados necessários para análise de qualidade de uma safra de vinho, e responderá se ela é de alta qualidade e, portanto, uma candidata a ser submetida a avaliação de especialista, ou se ela é de má qualidade e não deveria ser analisada.
